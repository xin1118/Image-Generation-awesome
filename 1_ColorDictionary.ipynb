```python
from IPython.display import display
from PIL import Image, ImageDraw, ImageFont
import io

# 定义颜色和描述信息（英文）
colors_with_descriptions = [
    {"name": "General Laboratory Area", "color": (255, 0, 0), "description": "A laboratory area suitable for multi-disciplinary experiments primarily conducted at lab benches."},
    {"name": "Dedicated Laboratory Area", "color": (0, 255, 0), "description": "Laboratory area with specific environmental requirements or specialized equipment such as electron microscopes and high-precision balances."},
    {"name": "Scientific Research Office Area", "color": (0, 0, 255), "description": "An area mainly used for office space to conduct research activities, including study rooms, academic activity rooms, etc."},
    {"name": "Scientific Research Exhibition Area", "color": (255, 255, 0), "description": "Space for displaying research and experimental results, typically in independent exhibition halls or galleries."},
    {"name": "Scientific Research Instructional Area", "color": (0, 255, 255), "description": "An area focused on teaching experiments, which can be divided into teaching labs, auxiliary rooms, etc."},
    {"name": "Scientific Research Testing Area", "color": (255, 0, 255), "description": "A testing area for industrial products or materials, composed of test rooms, auxiliary rooms, and public facilities."},
    {"name": "Wild Scientific Research Station", "color": (165, 42, 42), "description": "A research station for field observations in ecology, special environments, and atmospheric background."}
]

# 设置色卡参数
block_width = 300
card_height = 150
text_padding = 10
font_size = 20

# 创建空白图像，宽度为所有区块宽度之和加一些间距
img_width = len(colors_with_descriptions) * block_width + 50
img = Image.new('RGB', (img_width, card_height + 50), color='white')
draw = ImageDraw.Draw(img)


# 绘制色卡
x_offset = 25
for item in colors_with_descriptions:
    color = item["color"]
    name = item["name"]
    description = item["description"]

    # 绘制颜色块
    draw.rectangle([x_offset, 20, x_offset + block_width - text_padding, 20 + 70], fill=color)

    # 绘制名称和描述文本
    text_x = x_offset + text_padding
    text_y = 95
    draw.text((text_x, text_y), f"{name}:",  fill=(0, 0, 0))
    lines = [description[i:i+40] for i in range(0, len(description), 40)]
    for line in lines:
        text_y += font_size + 5
        draw.text((text_x, text_y), line, fill=(0, 0, 0))

    x_offset += block_width

# 将图像转换为字节流，以便在Jupyter Notebook中显示
with io.BytesIO() as output:
    img.save(output, format="PNG")
    display(Image.open(output))
```
